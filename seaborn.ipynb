# plot or total_bill

```python
import matplotlib.pyplot as plt

plt.plot(data['total_bill'])
plt.show()
```

# lineplot

```python
sns.lineplot(data['total_bill'])  #line plot of total_bill
```

# Scatterplot

```python
sns.scatterplot (x='total_bill' , y='tip',data=data) # we are doing scatter plot of total_bill and tip
```

```python
sns.histplot(data['total_bill'])
```

```python
sns.relplot(x='total_bill' ,y ='tip',hue ='time',style='smoker',data=data,kind='scatter',col='smoker')  #we are doing scatter plot of total_bill and tip
```

```python
sns.relplot(x='total_bill' ,y ='tip',hue ='time',data=data,kind='scatter',style='smoker',col='smoker',row='time')  #we are doing scatter plot of total_bill and tip
```

```python
sns.relplot(x='total_bill' ,y ='tip',data=data)
plt.title("total_bill vs tip")
plt.xlabel('total bill')
plt.ylabel('tip')
plt.show()
```

```python
import seaborn as sns

sns.relplot(x='total_bill' ,y = 'tip',data=data)
plt.title("total_bill vs tip")
plt.xlabel('total bill',fontsize=30,color='red')
plt.ylabel('tip',fontsize=30,color='red')
plt.show()
```
