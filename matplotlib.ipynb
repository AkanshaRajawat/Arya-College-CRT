# cmap= nipy_spectral

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.array([0,1,2,3,2])
y=np.array([0,8,2,6,3])
size=np.array([100,200,80,40,45])
c=np.array(['red','green','blue','yellow','black'])
plt.scatter(x,y,s=size,alpha=0.8,c=c,cmap="nipy_spectral")  #alpha is used the cmp=virdis,nipy_spectral
plt.colorbar()
plt.show()
```

# Explode 

```python
import matplotlib.pyplot as plt

y = [1,2,3,4]  # Example data for pie chart
mylabels = ["apple", "mango", "banana", "orange"]  # Corrected variable name
myexplode = [0.2, 0, 0, 0]  # Exploding the first slice

plt.pie(y, labels=mylabels, startangle=90, explode=myexplode)
plt.show()
```

# Shadow

```python
y=np.array([1,2,3,4])
mylabel= ["apple","mango","banana","orange"]
myexplode= [0.2,0,0,0.9]
plt.pie(y,labels=mylabel,startangle=90,explode=myexplode,shadow=True)  # shadow for 2d print
plt.show()
```

# autopct

```python
y=np.array([1,8,2,6])
mylabel=["co2","n2","o2","caco3"]
myexplode=[0.2,0,0,0]
mycolor=['red','orange','pink','yellow']
plt.pie(y,labels=mylabel,startangle=90,explode=myexplode,shadow=True,colors=mycolor,autopct='%1.1f%%')
plt.show()
```

# Legend
niche wala

```python
y=np.array([1,8,2,6])
mylabel=["co2","n2","o2","caco3"]
myexplode=[0.2,0,0,0]
mycolor=['red','orange','pink','yellow']
plt.pie(y,labels=mylabel,startangle=90,explode=myexplode,shadow=True,colors=mycolor,autopct='%1.1f%%')
plt.legend(title='color')
plt.show()
```
